```python
import pandas as pd
```

```python
ranking_df = pd.read_html('https://www.eliteleague.co.uk/stats/teams?id_season=36&id_stage=1')
```

```python
ranking_df[0].head(1)
```

```python
ranking_df[0].info()
```

```python
ranking_df1 = ranking_df[0].drop(['OTW','OTL/SOL','GF','GA','PTS %','PP%','PK%','SF Av.','SA Av.','GFA','GAA','PIMG','AVG ATT'],axis=1)
```

```python
ranking_df1
```

```python
ranking_df2 = ranking_df1.rename(columns={'Team':'Team Ranking','GP':'Games Played','PTS':'Points','W':'Wins','L':'Losses'})
```

```python
ranking_df2
```

```python
exported_values = list(ranking_df2.T.to_dict().values() )
```

```python
import json
```

```python
open('ranking','w').write( json.dumps(exported_values) )
```
